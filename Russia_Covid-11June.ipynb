## Lets take a look of death situation in the capital city of Moscow !!

```python
moscow_death=pd.read_csv('H:\\Assignment_GLE\\Inclass-2\\moscow_deaths.csv')
moscow_death.tail()
```

```python
t=moscow_death.groupby(by='Year')
w=t.sum()
Year_death=w.drop('Month',axis=1)
Year_death.sort_values(by='Deaths',ascending=False).plot(kind='bar',color='red')
plt.show()
```

## Interesting to note that in Moscow,the deaths in 2020 till 19th June2020,has been the lowest .
## This can be said that Covid19 is not proving to be as major death hurdle for the Russians as they have till now significantly lowered the risk.
## Most deaths were in 2010,2016 which means that Maybe it is because of Moscow metro bombings & Wildfires which resulted in more deaths in the city.
## Also,maybe exposure to severe cold to the lower class people have proven deadly which now is less because of lockdowns & increase in Govt. care for the Citizens

```python
j=moscow_death.groupby(by='Month')
q=j.sum()
Month_deaths=q.drop('Year',axis=1)
Month_deaths.sort_values(by='Deaths',ascending=False).plot(kind='bar',color='green')
plt.show()
```

## Moscow has the most deaths in Winter season January Month due to temperatures less than 0 degree celsius.
## Moscow has the least deaths in Summer season June Month due to  more temperatures
